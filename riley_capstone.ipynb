```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
namus = pd.read_csv('./data/used/namus_clean.csv', on_bad_lines = 'warn')
namus
```

```python
namus['DLC'] = pd.to_datetime(namus['DLC'])
namus['Date Modified'] = pd.to_datetime(namus['Date Modified'])
namus
```

```python
namus['Missing Age'] = namus['Missing Age'].str.extract(r'^(\d{2})', expand = False)
```

```python
namus['Race/Ethn 5'] = namus['Unnamed: 18']
```

```python
namus = namus.drop(['Legal Last Name', 'Legal First Name', 'Unnamed: 13', 'Unnamed: 18'], axis=1)
```

```python
namus
```

```python
namus = namus.rename(columns={'Legal Last Name CRCT':'Last Name', 'Legal First Name CRCT':'First Name', 'Race / Ethnicity 1':'Race/Ethn 1'})
```

```python
namus
```

```python
namus2020 = namus.loc[namus['DLC'].dt.year == 2020]
namus2020
```

```python
namus2020.State.value_counts()
```

```python
pops = pd.read_csv('./data/used/apportionment.csv')
```

```python
pops
```

```python
pops20 = pops.loc[pops['Year'] == 2020]
pops20
```

```python
pops20state = pops20[pops20['Geography Type'] == 'State']
pops20state
```

```python
pops20state = pops20state.drop(columns = ['Average Apportionment Population Per Representative', 'Change in Number of Representatives', 'Number of Representatives', 'Geography Type'])
```

```python
pops20state
```

```python
namus2020
```

```python
states = pd.read_csv('./data/used/states.csv')
states['Name'] = states.Name.str.strip()
```

```python
namus2020['State'] = namus2020.State.str.strip()
```

```python
pops20state = pd.merge(pops20state, states, left_on='Name', right_on='Name')
pops20state
```

```python
pops20state['Abbreviation'] = pops20state.Abbreviation.str.strip()
```

```python
namuspop2020 = pd.merge(namus2020, pops20state, left_on='State', right_on='Abbreviation')
```

```python
namuspop2020 = namuspop2020.drop(['Abbreviation', 'Percent Change in Resident Population'], axis=1)
```

```python
namuspop2020 = namuspop2020.rename(columns={"State": "State Abbreviation", "Name": "State"})
```

```python
namuspop2020
```

```python
statecases = namuspop2020.State.value_counts().to_frame()
statecases = statecases.reset_index()
```

```python
states_norm = pd.merge(statecases, pops20state, left_on='State', right_on='Name')
states_norm.head()
```

```python
states_norm['Resident Population'] = states_norm['Resident Population'].str.replace(',','').astype(int)
states_norm
```

```python
states_norm['Cases per 100K'] = (states_norm['count']/states_norm['Resident Population']*100000)
```
